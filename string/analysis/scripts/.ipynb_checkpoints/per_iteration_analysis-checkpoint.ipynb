# Analysis of confout.pdb files

This is to check progress of the string simulations in a "not CV" space. IE we can look at gate COM, salt bridge progress, sugar coordination etc. 

The purpose of this is to check integrity of the simulations. Ideally we will compare 2 - 4 confout files at a time. This way you can see if there is some average drift or instabilities happening potentially

#### input:
Use simulations parsed by `combine_confout.sh` which will take the `.gro` files from the `md/(#iteration)/restrained/confout.gro` and save only the protein. Then concatenate each bead output from this to have one continuous 'trajectory' which is # beads long

```python
import numpy as np
import MDAnalysis as mda
import matplotlib.pyplot as plt
from gate_functions import *
```

```python
#TODO fix this to be for new sims , right now we are just testing code 
indir = '../confout_files/measure_per_iteration'
traj = 'influx_apo_gate_CV'

#                 'influx_apo_gate_CV':[0,10,38,63,95,100,105,110,115,120,125,130,135,150,155,180,200,230,250]}

#here you can put what iterations #'s you have run
iterations_dir = {'efflux_apo_gate_CV':[0,10,38],\
                 'influx_apo_gate_CV':[0,10,50,100,150,200,250,300,350]}

universe_list = []
for iteration in iterations_dir[traj]:
    universe_list.append(mda.Universe('%s/%s.%i.string.pdb' %(indir, traj, iteration) ))
    
n_beads = 16 - 1
```

# Comparing bead RMSD to original homology models

```python
modeldir = '../../../steered/analysis/ref_models'

model_names = ["OUT.equilib.200ns", "OUT-OCC.start", "OCC.start", "IN-OCC.start", "IN.start"]
color_l = ["red", "blue", "green", "orange", "pink"]

model_l = []
for model in model_names:
    model_l.append(mda.Universe(f"{modeldir}/{model}.gro"))
```

```python
import MDAnalysis.analysis.rms

def calc_model_RMSD(model, traj):
    RMSD = MDAnalysis.analysis.rms.RMSD(reference=model.select_atoms("protein"), \
                                        atomgroup = traj.select_atoms("protein"), \
                                        center = True, select="backbone", \
            groupselections=["backbone and (resid 1-51 or resid 265-300 or resid 118-146 or resid 364-395)",\
                             "backbone and (resid 1-205 or resid 265-455)",\
                             "resid 30-37 or resid 289-295 or resid 136-145 or resid 386-394"])  #3,4,5 on rmsd

    RMSD.run(0)
    rmsd = RMSD.rmsd.T
    return rmsd
```

```python
traj_rmsd = {}
## shape of each group is (# iterations, size of RMSD.rmsd.T (5 for now), and # beads)
for name, model in zip(model_names, model_l):
    rmsd_l = []
    for iteration in universe_list:
        rmsd_l.append(calc_model_RMSD(model = model, traj = iteration))
    traj_rmsd[name] = rmsd_l
```

```python
fig, axs = plt.subplots(nrows = len(traj_rmsd), figsize = (10,10))
cmap_vals = plt.cm.get_cmap('viridis')
color_vals = cmap_vals(np.linspace(0,1,len(label_list)))

ylim = (0,6)

for ax_loc, rmsd_label in enumerate(traj_rmsd):
    # get out results, and then loop through them
    rmsd = np.array(traj_rmsd[rmsd_label])
   
    # plot each iteration
    for n, iteration in enumerate(rmsd):
        axs[ax_loc].plot(iteration[1], iteration[2], color = color_vals[n], label = label_list[n])
    
    axs[ax_loc].set_title(f"RMSD to {rmsd_label.lower().split('.')[0]} model")  #just make the name a bit prettier
    axs[ax_loc].set_ylim(ylim)
    axs[ax_loc].set_xlabel("Bead")
    axs[ax_loc].set_ylabel("RMSD A")
    

handles, labels = axs[ax_loc].get_legend_handles_labels()
fig.legend(handles, labels, loc='center right')

plt.tight_layout()

plt.xlim(0,15)
```

### Can also assess std of each bead to see if there are problem regions or highly variable states

Here I also make a dot for the gate COM of the initial atomistic trajectories (which were targets in the steered MD) and plot it ***APPROXIMATELY*** at the position in the trajectory where it corresponds to

```python
refs = ['IN.start.gro', 'IN-OCC.start.gro', 'OCC.start.gro', 'OUT-OCC.start.gro', 'OUT.equilib.200ns.gro']
if 'influx' in traj:
    refs = refs[::-1]   # if we are looking at influx sims, refs needs to be in the OPPOSITE order!! 

ref_universe = []
for ref in refs:
    ref_universe.append(mda.Universe('../../../steered/analysis/ref_models/%s' %ref))
    
ref_EC = []
ref_IC = []

for simulation in ref_universe:
    EC, IC = make_gate_arr(simulation, gate_EC = [(30,37), (289,295)], gate_IC = [(136,145), (386,394)])
    ref_EC.append(EC)
    ref_IC.append(IC)
```

```python
figsize = (13, 6)
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True, figsize = figsize)
markersize = 80

EC_std = np.std(np.array(EC_list), axis = 0)
IC_std = np.std(np.array(IC_list), axis = 0)

EC_mean = np.mean(np.array(EC_list), axis = 0)
IC_mean = np.mean(np.array(IC_list), axis = 0)

ref_locs = np.linspace(0, n_beads, num = len(ref_EC))

alpha = 0.3

ax1.plot(EC_mean, color = 'red')
ax1.fill_between(range(len(EC_mean)), (EC_mean + EC_std), (EC_mean - EC_std), alpha = alpha, color = 'orange')
ax1.scatter(ref_locs, ref_EC, color = 'orange', s = markersize, edgecolor = 'black')
ax1.set_title("Extracellular gate")
ax1.set_ylabel('Distance (A)')


ax2.plot(IC_mean, color = 'black')
ax2.fill_between(range(len(IC_mean)), (IC_mean + IC_std), (IC_mean - IC_std), alpha = alpha, color = 'blue')
ax2.scatter(ref_locs, ref_IC, color = 'blue', s = markersize, edgecolor = 'black')
ax2.set_title("Intracellular gate")
ax2.set_xlabel('Bead #')
ax2.set_ylabel('Distance (A)')

plt.tight_layout()
```
